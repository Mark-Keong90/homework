## Basic Level

### Part 1: Read in the file with `csv.reader()` and store it in an object called `file_nested_list`.

Hint: This is a TSV (tab-separated value) file, and `csv.reader()` needs to be told [how to handle it](https://docs.python.org/2/library/csv.html).

```python
import csv
import pandas as pd

from collections import namedtuple   # Convenient to store the data rows

DATA_FILE = 'C:/Users/Mark Keong/Documents/GitHub/classes/unit-2_project/data/chipotle_2.tsv'
```

```python
# Specify that the delimiter is a tab character
with open(DATA_FILE, mode='r') as f:
    chipotle_csv = [row for row in csv.reader(f, delimiter='\t')]
```

### Part 2: Separate `file_nested_list` into the `header` and the `data`.


```python
type(chipotle_csv)
```

```python
header = chipotle_csv[0]
data = chipotle_csv[1:]
```

```python
header
```

```python
df = pd.DataFrame(data=data, columns=header)
df
```

```python
df['item_price'].replace({'\$': ''}, regex= True, inplace=True)
```

```python
df.info()
```

```python
df[['quantity', 'item_price']] = df[['quantity', 'item_price']].astype(float, copy=False)
```

## Advanced Level


### Part 5: Calculate the average number of toppings per burrito.

Note: Let's ignore the `quantity` column to simplify this task.

Hint: Think carefully about the easiest way to count the number of toppings!


```python
df.head()
```

```python
df_burrito = df[df['item_name'].str.contains('Burrito')].reset_index()
df_burrito.drop(axis=1, columns='index')
```

```python
df_burrito.item_name.unique()
```

```python
df_burrito['item_count'] = df_burrito['choice_description'].str.count(',') + 1
```

```python
df_burrito.groupby('item_name')['item_count'].mean()
```

### Part 6: Create a dictionary. Let the keys represent chip orders and the values represent the total number of orders.

Expected output: `{'Chips and Roasted Chili-Corn Salsa': 18, ... }`
